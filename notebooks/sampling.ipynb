```python
import pandas as pd
import os

def create_random_sample(input_filepath, output_filepath, fraction=0.05):
    """
    Loads a large CSV, takes a purely random sample, and saves it to a new file.
    """
    print(f"Loading massive dataset from {input_filepath}...")
    df = pd.read_csv(input_filepath)
    
    print(f"Taking a random {fraction * 100}% sample...")
    # .sample() grabs random rows across the entire dataset
    # random_state=42 ensures you get the exact same random rows every time you run it
    sampled_df = df.sample(frac=fraction, random_state=42)
    
    # Ensure the output directory exists
    output_dir = os.path.dirname(output_filepath)
    if output_dir:
        os.makedirs(output_dir, exist_ok=True)
        
    # Save the new, smaller dataset without the index column
    sampled_df.to_csv(output_filepath, index=False)
    print(f"Success! Saved {len(sampled_df)} rows to {output_filepath}")

# ==========================================
# Execution
# ==========================================

input_file = '../data/raw/TEP_Faulty_Training.csv'
output_file = '../data/processed/TEP_Random_Sample_5_percent.csv'

# Run it to grab 5% of the data
create_random_sample(
    input_filepath=input_file, 
    output_filepath=output_file, 
    fraction=0.05
)
```
